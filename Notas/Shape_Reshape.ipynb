# Shape y Reshape

- *Shape : Muestra* forma que tiene el array. 
- **Reshape : Cambia** la forma del array

***Nota***: Deben tener la misma cantidad de valores.

```python
import numpy as np
```

```python
arr = np.random.randint(1,10,(3,2))
arr
```

```python
arr.shape # Muestra las dimensiones del array en este caso 3x2 que es la que definimos en la parte de arriba.
```

```python
np.reshape(arr,(2,3)) #(np-array, dimension)
# Modifica la dimension del array para adaptarse al par√°metro
```

```python
arr
```

```python
np.reshape(arr,(6,1),"C") 
# Reorganiza el array de acuerdo a la forma del Lenguaje C 
# Todas las Filas hasta terminar y sigue con la otra fila.
```

```python
np.reshape(arr,(6,1),"F")
# Reorganiza el array de acuerdo a la forma del Lenguaje Fortran
# Toda la columna y despu√©s continua con la otra columna.
```

```python
np.reshape(arr,(6,1),"A")
# Reorganiza el array de acuerdo a como sea mas optimizado para nuestro sistema. 
# Puede ser C o F. 
```

## Reto

- Crear un array de cualquier dimension y cambiar sus dimensiones. 
- Intenta cambiar el array de forma que no respete la estructura original

```python
arr3 = np.eye(3)
arr3
```

### Cambiar sus dimensiones

```python
np.reshape(arr3,(1,9))
# En el caso de la matriz identidad de igual si es F o C sera la misma. 
```

### Cambiar a forma que no respete su estructura original

```python
np.reshape(arr3,(2,2))
```
