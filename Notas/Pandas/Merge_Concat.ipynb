```python
import pandas as pd
import numpy as np
```

# Concat 

```python
df1 = pd.DataFrame({'A':['A0',"A1",'A2','A3'],
 'B':['B0','B1','B2','B3'],
 'C':['C0',"C1",'C2','C3'],
 'D':['D0','D1','D2','D3']})

df2 = pd.DataFrame({'A':['A4',"A5",'A6','A7'],
 'B':['B4','B5','B6','B7'],
 'C':['C4',"C5",'C6','C7'],
 'D':['D4','D5','D6','D7']})
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],ignore_index=True)
```

```python
pd.concat([df1,df2],axis=1)
```

# Mergue

```python
izq = pd.DataFrame({
        'key':['k0','k1','k2','k3'],
        'A':['A0','A1','A2','A3'],
        'B':['B0','B1','B2','B3']
 })

der = pd.DataFrame({
        'key':['k0','k1','k2','k3'],
        'C':['C0','C1','C2','C3'],
        'D':['D0','D1','D2','D3']
 })
```

```python
izq.merge(der,on='key') # Merge bsico, con el mismo nombre para la columna. 
```

```python
izq = pd.DataFrame({
        'key':['k0','k1','k2','k3'],
        'A':['A0','A1','A2','A3'],
        'B':['B0','B1','B2','B3']
 })

der = pd.DataFrame({
        'key_2':['k0','k1','k2','k3'],
        'C':['C0','C1','C2','C3'],
        'D':['D0','D1','D2','D3']
 })
```

```python
izq.merge(der,left_on='key',right_on='key_2') # Merge en caso de que las columnas se llamen diferente. 
# Inner join default.
```

```python
izq = pd.DataFrame({
        'key':['k0','k1','k2','k3'],
        'A':['A0','A1','A2','A3'],
        'B':['B0','B1','B2','B3']
 })

der = pd.DataFrame({
        'key_2':['k0','k1','k2',np.nan],
        'C':['C0','C1','C2','C3'],
        'D':['D0','D1','D2','D3']
 })
```

```python
izq.merge(der,left_on='key',right_on='key_2', how='left') # How es a quien se le da mas importancia. 
```
