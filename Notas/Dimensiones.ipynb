# Dimensiones

En Numpy, tenemos la facilidad de poder manipular datos de diferentes maneras, como lo puede ser un solo dato, lista de python una hoja de calculo, una serie de tiempo o imágenes.

Algunas de las dimensiones de los datos son : 
* scalar: dim = 0 Un solo dato o valor
* vector: dim = 1 Listas de Python
* matriz: dim = 2 Hoja de cálculo
* tensor: dim > 3 Series de tiempo o Imágenes


Estos son unos ejemplos de como se pueden ver las  diferentes dimensiones de los datos en Numpy, 

## Generando los diferentes dimensiones de los datos

```python
import numpy as np
```

```python
scalar = np.array(42)
print(scalar)
scalar.ndim
```

```python
vector = np.array([1,2,3])
print(vector)
vector.ndim
```

```python
matriz = np.array([[1,2,3],[4,5,6]])
print(matriz)
matriz.ndim
```

```python
tensor= np.array([[[1,2,3],[4,5,6]],[[1,2,3],[4,5,6]]])
print(tensor)
tensor.ndim
```

## Agregar o eliminar dimensiones

```python
vector = np.array([1,2,3],ndmin=10)
print(vector)
vector.ndim
# Definiendo un vector con 10 dimensiones sin importar los datos ingresados.
```

```python
expand = np.array([1,2,3],axis=0)
print(expand)
expand.ndim
# Agrega una dimension mas. 
```

```python
print(vector,vector.ndim)
vector_2 = np.squeeze(vector)
print(vector_2,vector_2.ndim)
# El Método squeeze elimina las dimensione del array que no se están utilizando. 
```
