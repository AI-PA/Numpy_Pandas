# Mineria de Datos con dataset LEGO.

```python
import pandas as pd
```

```python
df_lego_sets = pd.read_csv("./DATA/lego_sets.csv")         # df
df_parent_themes = pd.read_csv("./DATA/parent_themes.csv") # theme 
```

```python
df_lego_sets.head(5)
```

```python
df_parent_themes.head(5)
```

```python
#Merge df themes con lego sets
df_merge = df_lego_sets.merge(df_parent_themes, left_on='theme_name',right_on='name')
df_merge.head(10)
df_merge.drop(columns='name_y', inplace=True)
df_merge.head(15)
```

```python
# Filtrando cuales tienen licencia. 
df_licensed = df_merge[df_merge['is_licensed']==True]
df_licensed
```

```python
df_merge[df_merge['set_num'].isnull()].shape
# En este caso la data ya se encuentra limpia, pero en el video se encuentra con valores duplicados y valores vacíos.
```

```python
# filtrado por Star Wars
star_wars = df_licensed[df_licensed['parent_theme']=='Star Wars']
the_force = int((star_wars.shape[0] / df_licensed.shape[0])*100)
print(the_force)
```

```python
# Agrupado por Year.
df_licensed_sorted = df_licensed.sort_values('year')
df_licensed_sorted['count'] = 1
df_groupby_licensed = df_licensed_sorted.groupby(['year','parent_theme']).sum().reset_index()
df_groupby_licensed.drop(columns=['set_num','name_x','theme_name'], inplace=True)
```

```python
max_df = df_groupby_licensed.sort_values('count',ascending=False).drop_duplicates(['year'])
max_df.sort_values('year',inplace=True)
max_df
```
