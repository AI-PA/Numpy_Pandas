```python
# Librerías Necesarias.
import pandas as pd
```

```python
# Importando archivos.
    # df = DataFrame
df_data_historica = pd.read_csv("./Data/fifa_worldcup_matches.csv")
df_fixture = pd.read_csv("./Data/fifa_worldcup_fixture.csv")
df_data_faltante = pd.read_csv("./Data/fifa_worldcup_missing_data.csv")
```

# Limpiando Data

## df_fixture

```python
df_fixture.head(5)
df_fixture.info()
df_fixture.describe()
```

```python
df_fixture['home']= df_fixture['home'].str.strip()
df_fixture['away']= df_fixture['away'].str.strip()
# .str.strip() <- Elimina los espacios en blanco al inicio o fin de una palabra.
```

## df_missing_data & df_historical_data

```python
df_data_faltante.isnull() # Detecta si hay valore vacíos
df_data_faltante['home'].isnull() # Identificamos los datos faltantes
df_data_faltante.dropna(inplace=True) # Eliminamos los datos faltantes
```

```python
df_data_faltante
```

```python
df_data_faltante[df_data_faltante['home'].isnull()]
# Comprobación de datos faltantes eliminados.
```

### Combinar dataframes

```python
df_data_hisorica = pd.concat([df_data_historica,df_data_faltante], ignore_index=True)
df_data_hisorica
```

```python
# Eliminar Duplicados
df_data_hisorica.drop_duplicates( inplace=True)

# Ordenar por año. 
df_data_hisorica.sort_values('year',inplace=True)
```

```python
df_data_hisorica
```

## df_historical_data

```python
# Filtrar por 2 condiciones.
df_data_hisorica[df_data_hisorica['home'].str.contains('Sweden') & df_data_hisorica['away'].str.contains('Austria')] 
```

```python
#Index de la Operación
index_eliminar = df_data_hisorica[df_data_hisorica['home'].str.contains('Sweden') & df_data_hisorica['away'].str.contains('Austria')] .index
```

```python
df_data_hisorica.drop(index=index_eliminar, inplace=True)
```

```python
# Expresiones regulares
# Valores que contengan texto aparte de números. 
df_data_hisorica[df_data_hisorica['score'].str.contains('[^\d–]')]
```

```python
# Remplazando valores para quitar el (a.e.t)
df_data_hisorica['score'] = df_data_hisorica['score'].str.replace('[^\d–]',"",regex=True)
```

```python
# Comprobación de valores quitados. 
df_data_hisorica[df_data_hisorica['score'].str.contains(' [^\d-]')]
```

```python
# Limpieza de valores. 
df_data_hisorica['home'] = df_data_hisorica['home'].str.strip()
df_data_hisorica['away'] = df_data_hisorica['away'].str.strip()
```

```python
# Dividiendo la columna score en Home Goals y Away Goals
df_data_hisorica[["HomeGoals","AwayGoals"]] = df_data_hisorica['score'].str.split('–',expand=True)
```

```python
df_data_hisorica
```

```python
# Eliminando la columna score
df_data_hisorica.drop('score',axis=1,inplace=True)
```

```python
# Renombrando columnas.
df_data_hisorica.rename(columns={'home':'HomeTeam','away':'AwayTeam','year':'Year'},inplace=True)
```

```python
# Comprobando el tipo de dato de la columna. 
df_data_hisorica.dtypes
```

```python
# Cambiando tipos de datos. 
df_data_hisorica = df_data_hisorica.astype({'HomeGoals':int,'AwayGoals':int,'Year':int})
```

```python
df_data_hisorica.dtypes
```

```python
# Agregando una columna. 
df_data_hisorica['TotalGoals'] = df_data_hisorica['HomeGoals'] +df_data_hisorica['AwayGoals']
```

```python
df_data_hisorica
```

## Exportando dataframes limpios

```python
df_data_hisorica.to_csv('./Data/ETL-Fifa-Worldcup-Matches.csv',index=False)
df_fixture.to_csv('./Data/ETL-Fifa-Worldcup-Fixture.csv',index=False)
```
